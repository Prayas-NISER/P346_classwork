### Factorial of N

```python
# Factorial of n
def factorial(n):
    n = float(n)
    if n >= 1 and float(n)==int(n):
        n = int(n)
        fact = 1
        for i in range(1,n+1):
            fact = fact*i
        return fact
    elif int(n) == 0:
        return 1
    else:
        print("Invalid Input!")
        return None
```

```python
# READING A MATRIX FROM A .txt FILE

def readMat(file_name):
    file=open(str(file_name),"r")
    M=[[]]
    temp=''
    i=0
 #what follows below is a just accessing the strings one by one, and and converting the file data to list
 #of lists format
    for j in file:
        for k in j:
            if k=='\n':
                M[i]+=[float(temp)]
                temp=''
                i+=1
                M+=[[]]
            elif k!=',' and k!='\n':
                temp+=k
            elif k==',' or k=='':
                M[i]+=[float(temp)]
                temp=''
    return(M)

#Function for creating matrix.
def newMat(dat,r,c):
        mat = []
        for i in range(r):
            temp = []
            for j in range(c):
                temp.append(0)
            mat.append(temp)         #Creating a null matrix.
        for i in range(r):
            for j in range(c):
                    mat[i][j] = float(dat[c*i+j])
        return mat

# Function for printing a matrix in proper format.
def printMat(mat):
    r = len(mat)
    c = len(mat[0])
    print("[",end="")
    for i in range(r):
        for j in range(c):
            temp = str(mat[i][j])
            if i!=0 and j == 0:
                print(" "+temp+"     ",end="")   
            elif j==c-1:
                print(temp.rstrip(),end="")
            elif j!=c-1:
                print(temp+"     ",end="")
        if i!= r-1:
            print("\n")
        else:
            print("]")
    print()

    
# Multiplication of matrices
def prodMat(A,B): 
    r1 = len(A)
    r2 = len(B)
    c1 = len(A[0])
    c2 = len(B[0])
    if c1 == r2:              #Checking eligibility for multiplication.
        ans = []
        for i in range(r1):
            temp = []
            for j in range(c2):
                temp.append('0')
            ans.append(temp)         #Creating a null answer matrix.
        for i in range(r1):
            for j in range(c2):
                temp = 0
                for k in range(c1):
                    temp = temp + float(A[i][k])*float(B[k][j])
                ans[i][j] = temp
        return ans
    else:
        print("Invalid input!")
        return None
 
 # Create a null matrix of order n.
def create0(r,c):
    dat = "0"*r*c
    I = newMat(dat,r,c)
    return I
    

# Create Identity matrix of order n.
def createI(n):
    dat = "0"*n**2
    I = newMat(dat,n,n)
    i=0
    for x in I:
        x[i] = 1
        i+=1
    return I

# Extracting leading submatrices
def leadingsubMat(A,n):    # n = No. of the submatrix, e.g. for [[0 1 2][3 4 5][6 7 8]], the 1st leading submatrix
    r = len(A)
    c = len(A[0])
    if n>r or n>c:
        print("Dimension of submatrix exceeds that of actual matrix")
        return None
    else:
        sub=newMat(("0"*n*n),n,n)  # reads [[0]], 2nd one reads [[0 1][3 4]] and the 3rd one reads [[0 1 2][3 4 5][6 7 8]].
        for i in range(n):
            for j in range(n):
                sub[i][j] = A[i][j]
        return sub

# Extracting a column from a given matrix and return it as a column vector
def extractCol(A,m):         # Extracts the mth column of A
    r = len(A)
    c = len(A[0])
    if m>r or m>c:
        print("Dimension of submatrix exceeds that of actual matrix")
        return None
    else:
        col = []
        for i in range(r):
            temp=[]
            temp.append(A[i][m-1])
            col.append(temp)
        return col    

# Extracting a row from a given matrix and return it as a column vector
def extractRow(A,m):         # Extracts the mth row of A
    r = len(A)
    c = len(A[0])
    if m>r or m>c:
        print("Dimension of submatrix exceeds that of actual matrix")
        return None
    else:
        col = create0(r,1)
        for i in range(r):
            col[i][0] = A[m-1][i]
        return col    
    
# R = k*R
def scalerow(R,k,n):
    for i in range(n):
        R[i] = k*float(R[i])
    return R

# R1 = R1 - R2
def deductrow(R1,R2,n):
    for i in range(n):
        R1[i] = R1[i] - R2[i]
    return R1

# Finding the maximum value of a column
def maxMat(A,i1,i2,c): # Checks between A(i1)(c) and A(i2)(c)
    max = 0
    for i in range(i1,i2+1):
        if A[i][c] > max:
            max = A[i][c]
    return max

# Swap two rows
def rowswap(A,i1,i2):
    temp = A[i1]
    A[i1] = A[i2]
    A[i2] = temp
    return A

# Partial Pivoting for Gauss-Jordan Elimination
def partialpivot(Ab,m):  # Ab is the Augmented matrix; m is the pivot index
    r = len(Ab)
    c = len(Ab[0])
    swaps = 0
    pivot = Ab[m][m]
    max = maxMat(Ab,m,r-1,m)
    if pivot!=0 and pivot==max:
        return Ab, swaps
    if m==(r-1):
        if pivot==0:
            Ab = None
            swaps = 0
        return Ab, swaps
    else:
        for j in range(m+1,r):
            if Ab[j][m]!=0 and Ab[j][m]==max:
                Ab = rowswap(Ab,m,j)
                swaps+=1
                pivot = Ab[m][m]        
        if pivot==0:
            swaps=0
            Ab=None
        return Ab, swaps
        
# Gauss-Jordan Elimination
def gaussj(Ab,str=None): # For checking the determinant user may type 'det' in place of str, or 'inv' for inverse.
    r = len(Ab)
    c = len(Ab[0])
    flag = 0
    det = 1
    for i in range(r):
        Ab, swaps = partialpivot(Ab,i)
        flag+=swaps
        if Ab is None and str is None:
            return("No unique solution exists for the given system.")
            exit()
        elif Ab is None and str=='det':
            return("Indeterminate value found/Partial pivoting failed.")
            exit()
        elif Ab is None and str=='inv':
            return("Indeterminate value found/Partial pivoting failed.")
            exit()
        else:    
            det = det*Ab[i][i]            # Stores the determinant
            ratio=1/Ab[i][i]
            Ab[i] = scalerow(Ab[i],ratio,c)    # R[i] = R[i]/pivot
            for j in range(0,r):
                if i!=j and Ab[j][i]!=0:      # element is already ZERO.
                    factor = Ab[j][i]
                    for k in range(i,c):
                        Ab[j][k]-= factor * Ab[i][k]
    
    Ab = roundMat(Ab,2)
    
    if str==None:
        print("The solns are:")
        print()
        for i in range(4):    
            print(Ab[i][4])
            print()
    elif str=='inv':
        return Ab
    elif str=='det':
        if det!=None:
            return det*(-1)**flag
        else:
            return det

# Determinant of a matrix (Using Gauss Jordan El.)
def detMat(A):
    r = len(A)
    c = len(A[0])
    if r==c:
        det = gaussj(A,'det')
        return det
    else:
        print("Enter a square matrix")
        return None

# Inverse of matrix using Gauss-Jordan El.
def InvMat(A):
    n = len(A)
    I = createI(n)
    b = 2*n
    M = newMat("0"*n*b,n,b)
    for i in range(n):
        for j in range(b):
            if j<n:
                M[i][j] = A[i][j]
            if j>=n:
                M[i][j] = I[i][j-n]
    M = gaussj(M,'inv')
    if M!="Indeterminate value found/Partial pivoting failed.":
        ans = createI(n)
        for i in range(n):
            for j in range(n,2*n):
                ans[i][j-n] = M[i][j]
        return ans
    else:
        return("Inverse doesn't exist/Partial pivoting failed.")

# For accepting a square matrix of order n from the user
def acceptMat(r,c):
    dat = "0"*r*c
    A = newMat(dat,r,c)
    for i in range(r):
        for j in range(c):
            A[i][j] = float(input(f"Enter A[{i}][{j}]: "))
    return A

# Rounding off the elements of a matrix correct upto 2 decimal places
def roundMat(A,k):
    r = len(A)
    c = len(A[0])
    # A, r, c = Matrix and its dimensions
    # k = no. of decimal places to round off to.
    for i in range(r):
        for j in range(c):
            A[i][j]=float(format(A[i][j],f'.{k}f'))
    return A


def augment(matrix, vector, to_right=False):
    """
    params: matrix, vector, to_right=False
    returns the augmented matrix
    """
    for i in range(len(matrix)):
        if to_right:
            for j in vector[i]:
                matrix[i].insert(0, j)
        else:
            for j in vector[i]:
                matrix[i].append(j)
    return matrix

# Pivotizing a Matrix
def pivotize(A):
    r = len(A)
    c = len(A[0])
    for x in range(r):
        A,s = partialpivot(A,x)
    return A


# LU decomposition using Doolittle algorithm
def doolittle(A):
    r = len(A)
    c = len(A[0])  
    for i in range(r):
        #This section will check if the decompostion is possible
        #And if required will use partial pivoting to attempt factorisation
        sub = leadingsubMat(A,i+1)    # Exctracting the ith leading submatrix
        if detMat(sub)=="Indeterminate value found/Partial pivoting failed." or detMat(sub)==0: # detMat(sub) returns the 
            pP,s = partialpivot(A,i)                                                            # Determinant of submatrix
            if pP==None:
                return("LU Decomposition not possible")
                exit()
         
    #This section caclulates the value of the elements in L and U       
    for i in range(1,r):                 #choosing row number
        for j in range(c):               #choosing column number
            if i<=j:                     #Calculating u_{ij}
                temp=0
                for k in range(i):
                    temp+=A[i][k]*A[k][j]
                A[i][j]=A[i][j]-temp     #replacing the original matrix with the calculated value
            elif i>j:                    #Calculating l_{ij}
                temp=0
                for k in range(j):
                    temp+=A[i][k]*A[k][j]
                A[i][j]=(A[i][j]-temp)/A[j][j]
    A = roundMat(A,3)
    return A


# LU decomposition using Crout's algorithm
def crout(A):
    r = len(A)
    c = len(A[0])
    for i in range(r):
        #This section will check if the decompostion is possible
        #And if required will use partial pivoting to attempt factorisation
        sub = leadingsubMat(A,i+1)    # Exctracting the ith leading submatrix
        if detMat(sub)=="Indeterminate value found/Partial pivoting failed.":
            return("LU decomposition not possible")
            exit()
        if detMat(sub)==0:    # gaussj(sub,i,j,'det') returns the Determinant of the submatrix
            pP,s = partialpivot(A,i) 
            if pP==None:
                return("LU Decomposition not possible")
                exit()
    #This section caclulates the value of the elements in L and U       
    for j in range(1,r):                 #choosing row number
        for i in range(c):               #choosing column number
            if i>=j:                     #Calculating u_{ij}
                temp=0
                for k in range(j):
                    temp+=A[i][k]*A[k][j]
                A[i][j]=A[i][j]-temp     #replacing the original matrix with the calculated value
            elif i<j:                    #Calculating l_{ij}
                temp=0
                for k in range(i):
                    temp+=A[i][k]*A[k][j]
                A[i][j]=(A[i][j]-temp)/A[i][i]
    A = roundMat(A,3)
    return A


# LU decomposition using Cholesky's method
def cholesky(A):
    import math            # Importing math lib only to check square root
    r = len(A)            
    c = len(A[0])         
    for k in range(r):
        #This section will check if the decompostion is possible
        #And if required, will use partial pivoting to attempt factorisation
        sub = leadingsubMat(A,k+1)
        if detMat(sub)=="Indeterminate value found/Partial pivoting failed.":
            return("LU decomposition not possible")
            exit()
        if detMat(sub)==0:    # gaussj(sub,i,j,'det') returns the Determinant of the submatrix
            pP,s = partialpivot(A,i) 
            if pP==None:
                return("LU Decomposition not possible")
                exit()
    #This section caclulates the value of the elements in L and U
    #we dont need to calculate every combination of i,j since the matrix is symmetric
    for i in range(r):
        for j in range(i,r):
            if i==j:                   #diagonal elements
                temp=0
                for k in range(i):
                    temp+=A[i][k]**2
                A[i][i]=math.sqrt(A[i][i]-temp)
            if i<j:                    #calculating off diagonal elements
                temp=0
                for k in range(i):
                    temp+=A[i][k]*A[k][j]
                A[i][j]=(A[i][j]-temp)/A[i][i]
                A[j][i]=A[i][j]         #due to symmetricity of matrix
    A = roundMat(A,3)
    return A

# Forward substitution for LU 
def forsubs(L,b):
    r = len(L)
    c = len(b[0])
    X = create0(r,c)
    for i in range(r):
        for k in range(c):
            s = 0
            for j in range(i):
                s += L[i][j]*X[j][k]
            X[i][k] = (b[i][k]-s)*(1/L[i][i])
    return X

# Backward substitution for LU 
def backsubs(U,b):
    r = len(U)
    c = len(b[0])
    X = create0(r,c)
    for i in range(r-1,-1,-1):
        for k in range(c):
            s = 0
            for j in range(r-1,i,-1):
                s += U[i][j]*X[j][k]
            X[i][k] = (b[i][k]-s)*(1/U[i][i])
    return X

# Solving system of equations using Doolittle's algorithm
def solveDoolittle(Ab):
    r = len(Ab)
    c = len(Ab[0])
    
    # Separating A and b from the Augmented matrix
    A = create0(r,r)
    b = create0(r,1)
    for i in range(r):
        b[i][0] = Ab[i][-1]
        for j in range(c-1):
            A[i][j] = Ab[i][j]   # A and b are now separated
    
    A2 = A          # Creating a duplicate of A
    
    A2 = doolittle(A2)    # Performing LU decomposition on A using doolittle()
    
    # Separating L and U
    if A2!=None:          # Decomposing A2 into L and U using Doolittle
        L = create0(r,r)
        U = create0(r,r)
        for i in range(r):
            for j in range(r):
                if j<i:
                    L[i][j] = A2[i][j]
                if j>=i:
                    U[i][j] = A2[i][j]    # Separating L and U from the L/U augmented form obtained from doolittle()
                    if j==i:
                        L[i][i] = 1       # Making diagonal elements of L to 1
        
        y = forsubs(L,b)        # Solving Ly=b
        x = backsubs(U,y)       # Finally solving Ux=y
        x = roundMat(x,3)       # Using custom round function to round the matrix to 3 decimals
        return(x)
    else:
        return("No unique soln exists/LU Decomposition failed")
        exit()

    
# Solving system of equations using Crout's algorithm
def solveCrout(Ab):
    r = len(Ab)
    c = len(Ab[0])
    
    # Separating A and b from the Augmented matrix
    A = create0(r,r)
    b = create0(r,1)
    for i in range(r):
        b[i][0] = Ab[i][-1]
        for j in range(c-1):
            A[i][j] = Ab[i][j]   # A and b are now separated
    
    A2 = A          # Creating a duplicate of A
    
    A2 = crout(A2)    # Performing LU decomposition on A using crout()
    
    # Separating L and U
    if A2!=None:          # Decomposing A2 into L and U using crout
        L = create0(r,r)
        U = create0(r,r)
        for i in range(r):
            for j in range(r):
                if j<=i:
                    L[i][j] = A2[i][j]
                    if j==i:
                        U[i][i] = 1       # Making diagonal elements of U to 1
                if j>i:
                    U[i][j] = A2[i][j]    # Separating L and U from the L/U augmented form obtained from crout()
                            
        y = forsubs(L,b)        # Solving Ly=b
        x = backsubs(U,y)       # Finally solving Ux=y
        x = roundMat(x,3)
        return(x)
    else:
        return("No unique soln exists/LU Decomposition failed")
        exit()


# Solving system of equations using Cholesky's algorithm
def solveCholesky(Ab):
    r = len(Ab)
    c = len(Ab[0])
    
    # Separating A and b from the Augmented matrix
    A = create0(r,r)
    b = create0(r,1)
    for i in range(r):
        b[i][0] = Ab[i][-1]
        for j in range(c-1):
            A[i][j] = Ab[i][j]   # A and b are now separated
    
    A2 = A          # Creating a duplicate of A
    
    A2 = cholesky(A2)    # Performing LU decomposition on A using crout()
    
    # Separating L and U
    if A2!=None:          # Decomposing A2 into L and U using crout
        L = create0(r,r)
        U = create0(r,r)
        for i in range(r):
            for j in range(r):
                if j<=i:
                    L[i][j] = A2[i][j]
                    if j==i:
                        U[i][i] = A2[i][i]       # Making diagonal elements of U equal to A[i][i] 
                if j>i:
                    U[i][j] = A2[i][j]    # Separating L and U from the L/U augmented form obtained from crout()
                            
        y = forsubs(L,b)        # Solving Ly=b
        x = backsubs(U,y)       # Finally solving Ux=y
        x = roundMat(x,3)       
        return(x)
    else:
        return("No unique soln exists/LU Decomposition failed")
        exit()

# Inverse of a matrix using LU (Doolittle algorithm) decomposition
def invMat(M):
    n = len(M)
    I = createI(n)
    A = doolittle(M)  # Performing LU decomposition on A using doolittle()
    
    if A=="LU Decomposition not possible":
        return("Inverse doesn't exist/LU decomposition not possible for the given matrix.")
    
    elif detMat(M)==0:
        return("Det = 0. Inverse does not exist.")
    
    else:
        # Separating L and U
        if A!=None:          # Decomposing A2 into L and U using Doolittle
            L = create0(n,n)
            U = create0(n,n)
            for i in range(n):
                for j in range(n):
                    if j<i:
                        L[i][j] = A[i][j]
                    if j>=i:
                        U[i][j] = A[i][j]    # Separating L and U from the L/U augmented form obtained from doolittle()
                        if j==i:
                            L[i][i] = 1       # Making diagonal elements of L to 1

            # We now have L and U separately stored, which can be used.
            inv = create0(n,n)
            for i in range(n):
                temp1 = extractCol(I,i+1)  # Exctracting columns of I one by one for solving.
                y = forsubs(L,temp1)      # Solving Ly = I[i] (ith column of I)
                x = backsubs(U,y)         # Solving Ux = A^(-1)[i] (ith column of A^(-1))
                x = roundMat(x,2)
                for j in range(n):
                    inv[i][j]=x[j][0]
            return inv
        else:
            return("No unique soln exists/LU Decomposition failed")
            exit()
```

## Complex Numbers

```python
# Modulus
def modComplex(r,i):
    mod = (r**2+i**2)**(1/2)
    return mod
    
# Conjugate
def conjugate(r,i):
    i = -i
    return r,i
    
# Addition
def addComplex(r1,i1,r2,i2):
    sum_r = r1+r2
    sum_i = i1+i2
    return sum_r, sum_i
    
# Product
def prodComplex(r1,i1,r2,i2):
    prod_r = r1*r2 - i1*i2
    prod_i = i1*r2 + r1*i2
    return prod_r, prod_i
    
# Division
def divComplex(r1,i1,r2,i2):
    div_r = r1*r2 + i1*i2
    div_i = i1*r2 - r1*i2
    mod = (r2**2+i2**2)
    div_r = div_r/mod
    div_i = div_i/mod
    return div_r, div_i

# Phase angle
def phaseComplex(r,i):
    import math
    theta=math.atan(abs(i/r))
    if r>0 and i>0:
        return theta
    elif r<0 and i>0:
        return (math.pi-theta)
    elif r>0 and i<0:
        return (-theta)
    elif r<0 and i<0:
        return (theta-math.pi)

        
# For printing a complex number in proper format
def printComplex(r,i):
    if i>=0:
        print(str(r)+" + "+str(i)+"i")
    else:
        print(str(r)+" - "+str(-i)+"i") 
```

### Bisection

```python
# Function for differentiating (only for 1st or second order differentiation)
def differentiator(f,n,x,h=10**(-5)):    # Numerical nth (n = 1 or 2) Differentiation of f(x)
    if int(n)==1:                        # at x; h is the tolerance
        df = round((f(x+h)-f(x))/h,2)
        return df
    elif int(n)==2:
        d2f = round((f(x+h)+f(x-h)-2*f(x))/(h**2),2)
        return d2f
    
## BRACKETING ##
def bracket(func,low,high,maxm=12):
    # maxm stores the maximum no. of iterations allowed.
    truth = func(low)*func(high)<=0
    i = 0
    if truth == False:
        while i <maxm and truth == False: 
            if abs(func(low)) < abs(func(high)):
                low = low - 0.5*(high-low)
                i = i+1
            elif abs(func(high)) < abs(func(low)):
                high = high + 0.5*(high-low)
                i = i+1
            truth = func(low)*func(high)<=0        
        if i ==12:        
            return "Bad bracket, try some other interval"
        else:
            return [low,high]  
    else:
        return [low,high]         

## BISECTION ## 
# Obtaining roots of a function using Bisection
def bisection(f,a,b,tol=10**(-7),max_itr=100):
# defining all blank lists for future storage
    Iterartion=[]
    Root_val=[]
    if f(a)*f(b)>0:
        print("Do proper bracketing")
    for i in range(max_itr):
        c = (a+b)/2
        if abs(b - a) < tol: #if a and b are too close to the solution
            continue
        if f(a) * f(c) > 0:
            a = c
        else:
            b = c
        # storing no, of iterations, root and absolute error at each iteration in respective lists
        Iterartion.append(i)
        Root_val.append(c)
        
    return c,Iterartion,Root_val
```

### Laguerre's method

```python
# Function for differentiating (only for 1st or second order differentiation)
def differentiator(f,n,x,h=10**(-5)):    # Numerical nth (n = 1 or 2) Differentiation of f(x)
    if int(n)==1:                        # at x; h is the tolerance
        df = round((f(x+h)-f(x))/h,2)
        return df
    elif int(n)==2:
        d2f = round((f(x+h)+f(x-h)-2*f(x))/(h**2),2)
        return d2f

def derivative(f, h=10 ** -4):
    """
    Returns a function for numerical derivative of f
    """
    def df_dx(x, h=h):
        return (f(x + h) - f(x - h)) / (2 * h)
    return df_dx

def polynomial(a):
    """Generates polynomial function from a list of coefficients"""
    return lambda x: sum(a_i * pow(x, i) for i, a_i in enumerate(a))


def synthetic_division(coefficients, divisor):
    coefficients = coefficients[::-1]
    quotient = []
    quotient.append(coefficients[0])
    for i in range(1, len(coefficients)):
        quotient.append(coefficients[i] + (divisor * quotient[-1]))
    return quotient[0:-1][::-1]

def lag(f, order, alpha, ϵ=10 ** -4):
    df_dx = derivative(f)
    d2f_dx2 = derivative(df_dx)

    alpha_old = alpha + 1  # just to start
    while abs(alpha_old - alpha) > ϵ:
        G = df_dx(alpha) / f(alpha)
        H = G * G - (d2f_dx2(alpha) / f(alpha))
        deno_last_term = ((order - 1) * (order * H - G * G)) ** (1 / 2)
        deno_sum = G + deno_last_term
        deno_diff = G - deno_last_term
        deno = deno_sum
        if abs(deno_diff) > abs(deno_sum):
            deno = deno_diff
        a = order / deno  # this is less readable but its faster than if-else
        alpha_old = alpha
        alpha -= a
    return alpha


def laguerre(coefficients,guess=1.0, ϵ=10 ** -4):
    """
    Parameters :
        - array of coefficients in ascending order of power
            3*x**2 + 2*x - 4  => [-4, 2, 1]
        - ϵ the tolerence
        - an optional guess
    """
    order = len(coefficients) - 1
    roots = []
    for i in range(order):
        f = polynomial(coefficients)
        root = lag(f, order, guess, ϵ)
        coefficients = synthetic_division(coefficients, root)
        order = len(coefficients) - 1
        roots.append(root)
    return roots
```

### Regula-Falsi method

```python
# Finding roots using Regula-Falsi / Falsification method
def regulaFalsi(f,a,b,tol=10**(-6),max_itr=100):
    #defining all blank lists for future storage
    Iterartion=[]
    Root_val=[]
    error=[]
    c=a #initiating a guess at one of the end points of the bracket
    for i in range(max_itr):
        prev_c=c #storing the previous value
        c = b- (((b - a)*f(b))/(f(b)-f(a)))
        if abs(b-a)<tol: #if a and b are too close to the solution
            continue
        if f(a)*f(c)>0:
            a=c
        else:
            b=c
        if abs(c - prev_c) > tol: #the difference between c and the previous c is the absolute error in this iteration
            #storing no, of iterations, root and absolute error at each iteration in respective lists
            Iterartion.append(i)
            Root_val.append(c)
            e=abs(Root_val[i]-Root_val[i-1])
            error.append(e)
        else: #if the error converges and no significant error is noticed
            return c,Iterartion,Root_val, error
```

### Newton-Raphson method

```python
#Solving Equation using Newton-Raphson method       
def newtonRaphson(f,b,tol=10**(-5),max_itr=100):
    # defining all blank lists for future storage
    Iterartion = []
    Root_val = []
    for i in range(max_itr):
        c = b - (f(b)/differentiator(f,1,b,10**(-3)))
        if abs(b-c)>tol:
            b = c
            # storing no, of iterations, root and absolute error at each iteration in respective lists
            Iterartion.append(i)
            Root_val.append(c)
        else:
            return c,Iterartion,Root_val
```

# NUMERICAL INTEGRATION

```python
''' MIDPOINT METHOD '''
#f = function to be integrated; [a,b] = Interval of integration; 
#N = No. of steps
def midpoint_int(f,a,b,N):
    h = float(b-a)/N
    value = 0
    for i in range(N): 
        value = value + f((a + h/2.0) + i*h)
    return h*value

''' TRAPEZOIDAL METHOD '''
def trapezoidal_int(f,a,b,N):
    h = float(b-a)/N
    ans = 0
    for i in range(N):
        ans = ans + f(a +i*h) + f(a +(i+1)*h)
    return h*ans/2

''' SIMPSON METHOD '''
def simpson_int(f,a,b,N):
    h = float(b-a)/N
    ans = f(a) + f(b)
    for i in range(N):
        x = a + i*h
        if i%2 == 0:
            ans = ans + 2*f(x)
        else:
            ans = ans + 4*f(x)
    return ans*h/3

''' MONTE CARLO INTEGRATION '''
def montecarlo_int(f,a,b,N):
    import random
    integral = 0
    val =0
    for i in range(N):
        x = a + (b-a)*random.random()
        val += f(x)
    integral = float(((b-a)*val)/N)    
    return integral

''' MONTE CARLO with plot '''
# Monte-Carlo integration followed by the plotting 
# of integral value and standard dev.
def montecarlo_with_plot(f,a,b,N):
    import random
    from matplotlib import pyplot as plt
    value3 = []
    iter = []
    devi = []
   
    for j in range(300):
        integral = 0
        val =0  # integral
        val1 = 0     # standard deviation
        for i in range(N):
            x = a + (b-a)*random.random()
            val += f(x)
            val1 += (f(x))**2  
                                       
        integral = float(((b-a)*val)/N) 
        sd =  sqrt((1/N)*(val1)- ((1/N)*(val))**2) 
        devi.append(sd)     
        value3.append(integral)
        iter.append(N)
        N += 10
    # NOTE that N is a random number which is set to increase in multiples of 10.
    print(f"Value of the integral calculated using Monte Carlo Integration = {montecarlo_int(f,a,b,N)}")
    a,b,c = value3,iter,devi
    
    plt.figure(0)
    plt.plot(b,a,"b-")   # plotting the integral value
    plt.xlabel("N")
    plt.ylabel("Integral value")

    plt.figure(1)
    plt.plot(b,c,"g-") # plotting the standard deviation
    plt.xlabel("N")
    plt.ylabel("Std. Deviation")
```

## ODE Solver

```python
## FORWARD EULER (Explicit) 
"""Solve y' = f(x,y), intial condition y(x_o) = y_o, [x_0,x_n] = Range of x to be solved"""
def forward_euler(fn,step_size,x_0,y_0,x_n):  
        X=[x_0]
        Y=[y_0]
        N=int(abs(x_0-x_n)/step_size)                               #number of points to be plotted
        for i in range(N):
            x=X[-1]+step_size
            y=Y[-1]+step_size*fn(X[-1],Y[-1])                       #using forward derivative formula
            X.append(x)
            Y.append(y)
        return X,Y
    
# EULER BACKWARD (Implicit)
def euler_backward(func,step_size,x_0,y_0,x_n,newtonraphson_guess = 5):
        X=[x_0]
        Y=[y_0]
        N=int(abs(x_0-x_n)/step_size)                               #number of points to be plotted
        for i in range(N):
            x=X[-1]+step_size
            func_raph = lambda y_raphson: Y[-1] + step_size*func(x,y_raphson) - y_raphson
            y_nr,b,c=newtonraphson(func_raph,newtonraphson_guess,10**-5)
            y=Y[-1]+step_size*func(x,y_nr)                       #using forward derivative formula
            X.append(x)
            Y.append(y)
        return X,Y

# predictor corrector
def predictor_corrector(func,step_size,x_0,y_0,x_n):  
        X=[x_0]
        Y=[y_0]
        N=int(abs(x_0-x_n)/step_size)                               #number of points to be plotted
        for i in range(N):
            x=X[-1]+step_size
            k1=step_size*func(X[-1],Y[-1])
            yp = Y[-1] + k1
            k2=step_size*func(x,yp)
            y_corrected =Y[-1] +(k1+k2)/2                          #using forward derivative formula
            X.append(x)
            Y.append(y_corrected)
        return X,Y

# RK2 method
"""Solve y' = f(x,y), intial condition y(x_o) = y_o, x_n is your choice"""
def RK2(fn,step_size,x_0,y_0,x_n):  
        X=[x_0]
        Y=[y_0]
        N=int(abs(x_0-x_n)/step_size)                               #number of points to be plotted
        for i in range(N):
            x=X[-1]+step_size/2
            k1 = step_size*fn(X[-1],Y[-1])
            k2 = step_size*fn(x,Y[-1]+k1/2)
            y=Y[-1]+k2                       #using forward derivative formula
            X.append(x)
            Y.append(y)
        return X,Y

## RK 4 method
"""Solve y' = f(x,y), intial condition y(x_o) = y_o, x_n is your choice"""
def RK4(fn,step_size,x_0,y_0,x_n):  
        X=[x_0]
        Y=[y_0]
        N=int(abs(x_0-x_n)/step_size)                               #number of points to be plotted
        for i in range(N):
            x=X[-1]+step_size/2
            x2=X[-1]+step_size
            k1 = step_size*fn(X[-1],Y[-1])
            k2 = step_size*fn(x,Y[-1]+k1/2)
            k3 = step_size*fn(x,Y[-1]+k2/2)
            k4 = step_size*fn(x2,Y[-1]+k3)
            y=Y[-1]+(k1+2*k2+2*k3+k4)*(step_size)/6                       #using forward derivative formula
            X.append(x)
            Y.append(y)
        return X,Y  


# RK4 method for solving system of differential equation with one independent variable
def RK4_general(F,x,Y_0,h=0.01,N=100):
    """
    input: F: column matrix containing all the functions
              e.g. [f1,f2......,fn]
                  where each fi takes input as fi([y_0,y_1.....y_n],x)
           x : initial value of the independent variable
           Y_0: column matrix containing all initial values of the dependent variables
              e.g. [[y_1_0],[y_2_0]........,[y_n_0]]
    output: list of lists containing the set of values that satisfy (1)
            e.g.  [[y_1_0,y_1_1,..............y_1_N],
                   [y_2_0,y_2_1,..............y_2_N],
                   .
                   .
                   [y_n_0,y_n_1,..............y_n_N],
                   [x_0,x_1,....................x_N]]
    
    """
    sol=Y_0+[[x]]
    n=len(sol)-1
    for _ in range(N):
       #K1=[h*f_i([k[-1] for k in sol[:-1]],sol[-1][-1]) for f_i in F]
       #print(sol)
       K1=[h*f_i([sol[k][-1] for k in range(n)],sol[-1][-1]) for f_i in F]
       K2=[h*f_i([sol[k][-1]+K1[k]/2 for k in range(n)],sol[-1][-1]+h/2) for f_i in F]
       K3=[h*f_i([sol[k][-1]+K2[k]/2 for k in range(n)],sol[-1][-1]+h/2) for f_i in F]
       K4=[h*f_i([sol[k][-1]+K3[k] for k in range(n)],sol[-1][-1]+h) for f_i in F]
       sol[-1].append(sol[-1][-1]+h)
       for i in range(n):
           sol[i].append(sol[i][-1]+(K1[i]+2*K2[i]+2*K3[i]+K4[i])/6)
    return(sol)

def forward_euler2(dydx, y0, x0, xf, step_size):
    """ Yields solution from x=x0 to x=xf """
    x = []
    y = []
    x.append(x0)
    y.append(y0)

    n = int((xf-x0)/step_size)      # no. of steps
    for i in range(n):
        x.append(x[i] + step_size)

    for i in range(n):
        y.append(y[i] + step_size * dydx(y[i], x[i]))

    return x, y

# Solves differential equation using Runge-Kutta method
def runge_kutta(d2ydx2, dydx, x0, y0, z0, xf, step_size):
    """ Yields solution from x=x0 to x=xf
    y(x0) = y0 & y'(x0) = z0
    z = dy/dx
    """
    x = []
    y = []
    z = []      # dy/dx
    x.append(x0)
    y.append(y0)
    z.append(z0)

    n = int((xf-x0)/step_size)      # no. of steps
    for i in range(n):
        x.append(x[i] + step_size)
        k1 = step_size * dydx(x[i], y[i], z[i])
        l1 = step_size * d2ydx2(x[i], y[i], z[i])
        k2 = step_size * dydx(x[i] + step_size/2, y[i] + k1/2, z[i] + l1/2)
        l2 = step_size * d2ydx2(x[i] + step_size/2, y[i] + k1/2, z[i] + l1/2)
        k3 = step_size * dydx(x[i] + step_size/2, y[i] + k2/2, z[i] + l2/2)
        l3 = step_size * d2ydx2(x[i] + step_size/2, y[i] + k2/2, z[i] + l2/2)
        k4 = step_size * dydx(x[i] + step_size, y[i] + k3, z[i] + l3)
        l4 = step_size * d2ydx2(x[i] + step_size, y[i] + k3, z[i] + l3)

        y.append(y[i] + (k1 + 2*k2 + 2*k3 + k4)/6)
        z.append(z[i] + (l1 + 2*l2 + 2*l3 + l4)/6)

    return x, y, z


# Returns the result of Lagrange's interpolation formula
def lagrange_interpolation(zeta_h, zeta_l, yh, yl, y):
    zeta = zeta_l + (zeta_h - zeta_l) * (y - yl)/(yh - yl)
    return zeta


# Solves 2nd order ODE given Dirichlet boundary conditions
def shooting_method(d2ydx2, dydx, x0, y0, xf, yf, z_guess1, z_guess2, step_size=0.05, tol=1e-6):
    '''x0: Lower boundary value of x
    y0 = y(x0)
    xf: Upper boundary value of x
    yf = y(xf)
    z = dy/dx
    '''
    x, y, z = runge_kutta(d2ydx2, dydx, x0, y0, z_guess1, xf, step_size)
    yn = y[-1]

    if abs(yn - yf) > tol:
        if yn < yf:
            zeta_l = z_guess1
            yl = yn

            x, y, z = runge_kutta(d2ydx2, dydx, x0, y0, z_guess2, xf, step_size)
            yn = y[-1]

            if yn > yf:
                zeta_h = z_guess2
                yh = yn

                # calculate zeta using Lagrange interpolation
                zeta = lagrange_interpolation(zeta_h, zeta_l, yh, yl, yf)

                # using this zeta to solve using RK4
                x, y, z = runge_kutta(d2ydx2, dydx, x0, y0, zeta, xf, step_size)
                return x, y, z

            else:
                print("Bracketing FAIL! Try another set of guesses.")


        elif yn > yf:
            zeta_h = z_guess1
            yh = yn

            x, y, z = runge_kutta(d2ydx2, dydx, x0, y0, z_guess2, xf, step_size)
            yn = y[-1]

            if yn < yf:
                zeta_l = z_guess2
                yl = yn

                # calculate zeta using Lagrange interpolation
                zeta = lagrange_interpolation(zeta_h, zeta_l, yh, yl, yf)

                x, y, z = runge_kutta(d2ydx2, dydx, x0, y0, zeta, xf, step_size)
                return x, y, z

            else:
                print("Bracketing FAIL! Try another set of guesses.")


    else:
        return x, y, z         # bang-on solution with z_guess1
```

```python
def read_csv(file):
    C=np.genfromtxt(file,delimiter=',')
    X=[]
    Y=[]
    for i in range(len(C)):
        X.append(C[i][0])
        Y.append(C[i][1])
    return X,Y

from math import e, sqrt, log as ln
# Linear fit
def linear_fit(xs, ys):
    # y = c + m * x
    mean_x = sum(xs) / len(xs)
    mean_y = sum(ys) / len(ys)
    m = sum((xs[i] - mean_x) * (ys[i] - mean_y) for i in range(len(xs))) / sum(
        (xs[i] - mean_x) ** 2 for i in range(len(xs))
    )
    c = mean_y - m * mean_x
    return c, m


# exponential fit
def exponential_fit(xs, ys):
    # y = a * e**(p * x)

    # transforming data into linear by taking log
    ys = list(map(ln, ys))

    # linear fit
    c, m = linear_fit(xs, ys)

    # transforming back to exponentials
    a = e ** c
    p = m
    return a, p


def pearsons_r(xs, ys):
    mean_x = sum(xs)/len(xs)
    mean_y = sum(ys)/len(ys)

    r = sqrt(sum((xs[i] - mean_x)*(ys[i] - mean_y) for i in range(len(xs)))**2 \
             /(sum((xs[i] - mean_x)**2 for i in range(len(xs)))* \
               sum((ys[i] - mean_y)**2 for i in range(len(ys)))))
    return r


def polynomial_fitting(X,Y, order):
    X1=copy.deepcopy(X)
    Y1=copy.deepcopy(Y)
    order+=1
    
    A=[[0 for j in range(order)] for i in range(order)]
    vector=[0 for i in range(order)]

    for i in range(order):
        for j in range(order):
            for k in range(len(X)):
                A[i][j] += X[k]**(i+j)

    for i in range(order):
        for k in range(len(X)):
            vector[i] += X[k]**i * Y[k]

    # Solution finding using LU decomposition using Doolittle's condition L[i][i]=1
    # partial pivoting to avoid division by zero at pivot place
    A, vector = partial_pivot_LU(A, vector, order)
    A = LU_doolittle(A,order)
    
    solution = for_back_subs_doolittle(A,order,vector)

    return solution[0:order]

        
def plot_graph_poly(X, Y, sol, order):
    yfit=[0 for i in range(len(X))]
    # finding yfit
    for k in range(len(X)):
        for l in range(order):
            yfit[k]+=sol[l]*X[k]**l
    
    # plotting X and y_fit
    plt.plot(X, yfit, 'r-', label="Curve fit with polynomial of degree = "+ str(order-1))


#partial pivot for poly fit
def partial_pivot(Ab,m,nrows):
    pivot = Ab[m][m]    # declaring the pivot
    if (Ab[m][m] != 0):
        return Ab    # return if partial pivot is not required
    else:
        for r in range(m+1,nrows):
            pivot=Ab[r][m]
            # check for non-zero pivot and swap rows with it
            for k in range(m+1,nrows):
                if abs(Ab[k][m])>pivot:
                    pivot=Ab[k][m]
                    r=k
            if Ab[r][m] != 0:
                pivot = Ab[r][m]
                Ab=swap_rows(Ab,m,r)
                return Ab
            else:
                r+=1
    if (pivot==0):    # no unique solution case
        return None

# LU do little for poly fit
def LU_doolittle(mat,n):
    for i in range(n):
        for j in range(n):
            if i>0 and i<=j: # changing values of upper triangular matrix
                sum=0
                for k in range(i):
                    sum+=mat[i][k]*mat[k][j]
                mat[i][j]=mat[i][j]-sum
            if i>j: # changing values of lower triangular matrix
                sum=0
                for k in range(j):
                    sum+=mat[i][k]*mat[k][j]
                mat[i][j]=(mat[i][j]-sum)/mat[j][j]
    return mat


# LU decomposition using Doolittle's condition L[i][i]=1
# without making separate L and U matrices

def LU_doolittle(mat,n):
    for i in range(n):
        for j in range(n):
            if i>0 and i<=j: # changing values of upper triangular matrix
                sum=0
                for k in range(i):
                    sum+=mat[i][k]*mat[k][j]
                mat[i][j]=mat[i][j]-sum
            if i>j: # changing values of lower triangular matrix
                sum=0
                for k in range(j):
                    sum+=mat[i][k]*mat[k][j]
                mat[i][j]=(mat[i][j]-sum)/mat[j][j]
    return mat


# Function to find the solution matrix provided a vector using 
# forward and backward substitution respectively

def for_back_subs_doolittle(mat,n,vect):
    # initialization
    y=[0 for i in range(n)]
    
    # forward substitution
    y[0]=vect[0]
    for i in range(n):
        sum=0
        for j in range(i):
            sum+=mat[i][j]*y[j]
        y[i]=vect[i]-sum
    
    # backward substitution
    x[n-1]=y[n-1]/mat[n-1][n-1]
    for i in range(n-1,-1,-1):
        sum=0
        for j in range(i+1,n):
            sum+=mat[i][j]*x[j]
        x[i]=(y[i]-sum)/mat[i][i]
    del(y)
    return x
        
def plot_graph_poly(X, Y, sol, order):
    
    # plot points in array X
    #plt.scatter(X, Y, s=50, color='blue')
    
    # finding yfit
    yfit=[0 for i in range(len(X))]
    for k in range(len(X)):
        for l in range(order):
            yfit[k]+=sol[l]*X[k]**l
    
    # plotting X and y_fit
    plt.plot(X, yfit, 'r-', label="Curve fit with polynomial of degree = "+ str(order-1))

def partial_pivot_LU (mat, vec, n):
    for i in range (n-1):
        if mat[i][i] ==0:
            for j in range (i+1,n):
                # checks for max absolute value and swaps rows 
                # of both the input matrix and the vector as well
                if abs(mat[j][i]) > abs(mat[i][i]):
                    mat[i], mat[j] = mat[j], mat[i]
                    vec[i], vec[j] = vec[j], vec[i]
    return mat, vec


def find_stats(X, Y):
    n=len(x)
    Sx=sum(X)
    Sy=sum(Y)

    x_mean=sum(X)/n
    y_mean=sum(Y)/n

    Sxx=0
    Sxy=0
    Syy=0
    for i in range(n):
        Sxx = Sxx + x[i]**2
        Sxy = Sxy + x[i]*y[i]
        Syy = Syy + y[i]**2
    
    return n, x_mean, y_mean, Sx, Sy, Sxx, Syy, Sxy

def Std_dev(X):
    var=0
    mean=Mean(X)
    for i in range(len(X)):
        var+=(X[i]-mean)**2
    var=var/len(X)
    std=np.sqrt(var)
    return std
```
